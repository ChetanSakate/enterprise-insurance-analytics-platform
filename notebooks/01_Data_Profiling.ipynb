```python
import pandas as pd

file_path = "../data/raw/Policy Data.xlsx"

customer_df = pd.read_excel(
    file_path,
    sheet_name="Customer Information"
)

policy_df = pd.read_excel(
    file_path,
    sheet_name="Policy Details"
)

claims_df = pd.read_excel(
    file_path,
    sheet_name="Claims"
)

payment_df = pd.read_excel(
    file_path,
    sheet_name="Payment History"
)

additional_df = pd.read_excel(
    file_path,
    sheet_name="Additional Fields"
)
```

```python
datasets = {
    "Customer": customer_df,
    "Policy": policy_df,
    "Claims": claims_df,
    "Payments": payment_df,
    "Additional": additional_df
}
```

```python
for name, data in datasets.items():

    print("\n" + "="*60)
    print(name)
    print("="*60)

    print("Rows :", data.shape[0])
    print("Columns :", data.shape[1])

    print("\nData Types")
    print(data.dtypes)

    print("\nMissing Values")
    print(data.isnull().sum())

    print("\nDuplicates")
    print(data.duplicated().sum())
```
