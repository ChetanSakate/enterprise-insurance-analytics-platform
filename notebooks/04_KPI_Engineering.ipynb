```python
import pandas as pd
import numpy as np

insurance_master = pd.read_csv(
    "../data/processed/insurance_master_dataset.csv"
)

claims_fact = pd.read_csv(
    "../data/processed/claims_fact.csv"
)

payment_fact = pd.read_csv(
    "../data/processed/payment_fact.csv"
)
```

```python
insurance_master["Policy Start Date"] = pd.to_datetime(
    insurance_master["Policy Start Date"]
)

insurance_master["Policy End Date"] = pd.to_datetime(
    insurance_master["Policy End Date"]
)

claims_fact["Date of Claim"] = pd.to_datetime(
    claims_fact["Date of Claim"]
)

claims_fact["Settlement Date"] = pd.to_datetime(
    claims_fact["Settlement Date"]
)

payment_fact["Date of Payment"] = pd.to_datetime(
    payment_fact["Date of Payment"]
)
```

```python
# Executive KPI Calculations

# KPI 1 Total Policies

total_policies = insurance_master["Policy ID"].nunique()

print(total_policies)
```

```python
# KPI 2 Total Customers

total_customers = insurance_master["Customer ID"].nunique()

print(total_customers)
```

```python
# KPI 3 Total Premium Revenue

total_premium = insurance_master[
    "Premium Amount"
].sum()

print(total_premium)
```

```python
# KPI 4 — Total Coverage Amount

total_coverage = insurance_master[
    "Coverage Amount"
].sum()

print(total_coverage)
```

```python
# KPI 5 — Total Claim Amount

total_claim_amount = claims_fact[
    "Claim Amount"
].sum()

print(total_claim_amount)
```

```python
# KPI 6 - Total Payment Collected

total_payment_collected = payment_fact[
    payment_fact["Payment Status"]=="Successful"
]["Amount Paid"].sum()

print(total_payment_collected)
```

```python
# KPI 7 - Active Policies

active_policies = insurance_master[
    insurance_master["Status"]=="Active"
].shape[0]

print(active_policies)
```

```python
# KPI 8 - Active Policy %

active_policy_pct = round(
    (active_policies / total_policies)*100,
    2
)

print(active_policy_pct)
```

```python
# KPI 9 - Policies Expiring This Year

current_year = pd.Timestamp.today().year

expiring_this_year = insurance_master[
    insurance_master["Policy End Date"].dt.year == current_year
].shape[0]

print(
    f"Policies Expiring This Year: {expiring_this_year}"
)
```

```python
# KPI 10 - Renewal Rate%

renewed = insurance_master[
    insurance_master["Renewal Status"]=="Renewed"
].shape[0]

renewal_rate = round(
    (renewed / total_policies)*100,
    2
)

print(renewal_rate)
```

```python
## Customer Analytics KPIs

# Age Bucket Wise Policies

age_bucket_kpi = insurance_master.groupby(
    "Age Bucket"
)["Policy ID"].count().reset_index()

age_bucket_kpi
```

```python
# Gender Wise Policies

gender_kpi = insurance_master.groupby(
    "Gender"
)["Policy ID"].count().reset_index()

gender_kpi
```

```python
# Occupation Wise Policies

occupation_kpi = insurance_master.groupby(
    "Occupation"
)["Policy ID"].count().reset_index()

occupation_kpi
```

```python
# Marital Status Wise Policies

marital_kpi = insurance_master.groupby(
    "Marital Status"
)["Policy ID"].count().reset_index()

marital_kpi
```

```python
## Policy Analytics KPIs

# Policy Type Wise Policy Count

policy_type_kpi = insurance_master.groupby(
    "Policy Type"
)["Policy ID"].count().reset_index()

policy_type_kpi
```

```python
# Policy Status Distribution

policy_status_kpi = insurance_master.groupby(
    "Status"
)["Policy ID"].count().reset_index()

policy_status_kpi
```

```python
# Premium by Policy Type

premium_policy_type = insurance_master.groupby(
    "Policy Type"
)["Premium Amount"].sum().reset_index()

premium_policy_type
```

```python
## Claims Analytics KPIs

#Claim Status Wise Count

claim_status_kpi = claims_fact.groupby(
    "Claim Status"
)["Claim ID"].count().reset_index()

claim_status_kpi
```

```python
# Claim Amount by Status

claim_amount_status = claims_fact.groupby(
    "Claim Status"
)["Claim Amount"].sum().reset_index()

claim_amount_status
```

```python
# Average Claim Amount

avg_claim_amount = round(
    claims_fact["Claim Amount"].mean(),
    2
)

print(avg_claim_amount)
```

```python
# Average Settlement Days

avg_settlement_days = round(
    claims_fact["Settlement Days"].mean(),
    2
)

print(avg_settlement_days)
```

```python
## Payment Analytics KPIs

# Payment Status Count

payment_status_kpi = payment_fact.groupby(
    "Payment Status"
)["Payment ID"].count().reset_index()

payment_status_kpi
```

```python
# Payment Method Analysis

payment_method_kpi = payment_fact.groupby(
    "Payment Method"
)["Payment ID"].count().reset_index()

payment_method_kpi
```

```python
# Payment Success Rate

successful_payments = payment_fact[
    payment_fact["Payment Status"]=="Successful"
].shape[0]

payment_success_rate = round(
    (successful_payments / len(payment_fact))*100,
    2
)

print(payment_success_rate)
```

```python
## Risk Analytics KPIs

# Risk Category Distribution

risk_kpi = insurance_master.groupby(
    "Risk Category"
)["Policy ID"].count().reset_index()

risk_kpi
```

```python
# High Risk Policies

high_risk_count = insurance_master[
    insurance_master["Risk Category"]=="High Risk"
].shape[0]

print(high_risk_count)
```

```python
# Risk vs Renewal

risk_renewal = pd.crosstab(
    insurance_master["Risk Category"],
    insurance_master["Renewal Status"]
)

risk_renewal
```

```python
## Executive Insurance KPIs

# Claim Ratio %

claim_ratio = round(
    (total_claim_amount /
     total_premium)*100,
    2
)

print(claim_ratio)
```

```python
# Average Premium Per Policy

avg_premium = round(
    total_premium /
    total_policies,
    2
)

print(avg_premium)
```

```python
# Average Coverage Per Policy

avg_coverage = round(
    total_coverage /
    total_policies,
    2
)

print(avg_coverage)
```

```python
## Executive KPI Summary Table

executive_kpis = pd.DataFrame({
    "KPI":[
        "Total Policies",
        "Total Customers",
        "Total Premium Revenue",
        "Total Claim Amount",
        "Claim Ratio %",
        "Active Policy %",
        "Renewal Rate %",
        "Payment Success Rate %"
    ],
    "Value":[
        total_policies,
        total_customers,
        total_premium,
        total_claim_amount,
        claim_ratio,
        active_policy_pct,
        renewal_rate,
        payment_success_rate
    ]
})

executive_kpis
```

```python
# Export KPI Tables

executive_kpis.to_excel(
    "../data/reports/kpi_summary.xlsx",
    index=False
)
```
